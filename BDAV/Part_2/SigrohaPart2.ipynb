```python
import sys
#checking the system python version for compatibility of NYTimesArticleAPI
print(sys.version)
```

```python
#after installing NYTimesArticleAPI using pip install nytimesarticle
#importing ArticleAPI
from nytimesarticle import articleAPI
#using the API Key fetched form developer.nytimes.com
api = articleAPI ('e7190054cbac48baaf1f03679da0cf44')
#taking a list to save the urls
urlList = ['']
#loop to fetch articles from pages for the keyword "Gun Violence"
for i in range (0,250): 
    articles = api.search (q = 'Gun Violence', begin_date = 20170701, page=i)
    try:
        for art in articles['response']['docs']:
            #priting th web_url so obtained form previous step
            print(art['web_url'])
            #saving it in url_list
            urlList.append(art['web_url'])
    except:
        s = "Exception encountered"
#to know the number of urls so accumulated
print(len(urlList)) 
```

```python
#to extract html pages from URLs we are using urllib feature & for the text from the articles on html pages we are using BeautifulSoup
#request urllib in program
import urllib.request
from bs4 import BeautifulSoup
#using an empty string to save the text data
file_data = ''
for urls in urlList:
   try:
       response = urllib.request.urlopen(urls)
       soup = BeautifulSoup(response,"html.parser")
#loop to find the paragraphs using the <p> tags
       for p in soup.find_all('p'):
            file_data = file_data + "\n"
#saving in the strings
            file_data = file_data + p.get_text()
   except:
       s = "Exception encountered"
#to verify the length of urlList so as to assure all the urls previously obtained are scanned for <p> tags:-
print(len(urlList))
print(file_data)
#saving the data from the string in a txt file:-
with open("Output.txt", "w+") as text_file:
   text_file.write(file_data)
   
text_file.close()
```
